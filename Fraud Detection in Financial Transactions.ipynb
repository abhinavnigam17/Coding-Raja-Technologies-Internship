```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
credit_card_data = pd.read_csv("creditcard.csv")
```

## Distribution of Legit Transaction & Fraudulant Transaction

```python
credit_card_data['Class'].value_counts()
```

## The Dataset is Highly Unbalanced :-
#### Where 0 is Normal Transaction
#### And 1 is Fraudulent Transcation

## Seprating Data For Anlaysis

```python
legit = credit_card_data[credit_card_data.Class == 0 ]
fraud = credit_card_data[credit_card_data.Class == 1 ]
```

```python
print(legit.shape)
print(fraud.shape)
```

## Statistical Measures of Data

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

## Compare the value for Both Transcation

```python
credit_card_data.groupby('Class').mean()
```

## Under Sampling

#### Building a Sample Dataset containing similar distribution of Normal and fraudulent Transcation
###### Number of Fraudulent Transcation = 492

```python
legit_sample =legit.sample(n=492)
```

### Concatenating 2 Dataframe

```python
new_dataset = pd.concat([legit_sample, fraud],axis=0)
# axis=0 means Rows & axis=1 means columns 
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

### Spliting the data into Features & Targets

```python
X= new_dataset.drop(columns= 'Class', axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

### Split the data into Training Data and Testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

### Logistic Regression Model
 

```python
model = LogisticRegression()
```

#### Training the Logistic Regression Model with Training Data

```python
model.fit(X_train, Y_train)
```

### Model Evaluation & Accuracy Score

### Accuracy Score on Training Data

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training Data : ',training_data_accuracy)
```

### Accuracy on Test Data 

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print("Accuracy Score on Data : ",test_data_accuracy)
```
